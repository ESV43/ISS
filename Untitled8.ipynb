```python
# [CELL 2] - Imports
import requests
import numpy as np
import matplotlib.pyplot as plt
from skyfield.api import load, EarthSatellite, wgs84
from skyfield.timelib import Time
from datetime import datetime, timedelta
import json
import csv
from ipywidgets import widgets, interact
from IPython.display import display, clear_output
import pytz
```

```python
def fetch_iss_tle():
    """Fetch current ISS TLE data from CelesTrak"""
    try:
        # CelesTrak API endpoint for ISS
        url = "https://celestrak.org/NORAD/elements/gp.php?CATNR=25544&FORMAT=TLE"

        response = requests.get(url, timeout=10)
        response.raise_for_status()

        # Parse the TLE data
        lines = response.text.strip().split('\n')

        if len(lines) >= 3:
            name = lines[0].strip()
            line1 = lines[1].strip()
            line2 = lines[2].strip()

            # Validate TLE format
            if line1.startswith('1 ') and line2.startswith('2 '):
                print(f"Successfully fetched TLE for: {name}")
                print(f"Line 1: {line1}")
                print(f"Line 2: {line2}")
                return name, line1, line2
            else:
                raise ValueError("Invalid TLE format received from CelesTrak.")

    except requests.exceptions.RequestException as e:
        print(f"Error fetching TLE data: {e}")
        # Fallback TLE data (example from a recent date)
        name = "ISS (ZARYA)"
        line1 = "1 25544U 98067A   24213.50000000  .00016717  00000-0  30613-3 0  9993"
        line2 = "2 25544  51.6416 221.3612 0007804 117.4477 242.7448 15.49491294463378"
        print("Using fallback TLE data.")
        return name, line1, line2

# Fetch the TLE and store in global variables for other tasks
iss_name, tle_line1, tle_line2 = fetch_iss_tle()
```

```python
# [REVISED CODE FOR TASK 3 CELL]

def calculate_iss_passes(latitude, longitude, elevation=0, days_ahead=5, city_name=None):
    """Calculate ISS passes for a given location"""

    if city_name:
        print(f"\n--- Calculating ISS Passes for {city_name} ---")

    # Validate coordinates
    if not (-90 <= latitude <= 90):
        raise ValueError("Latitude must be between -90 and 90 degrees")
    if not (-180 <= longitude <= 180):
        raise ValueError("Longitude must be between -180 and 180 degrees")

    # Load timescale and create satellite object
    ts = load.timescale()
    # Assumes iss_name, tle_line1, tle_line2 are globally available from Task 2
    satellite = EarthSatellite(tle_line1, tle_line2, iss_name, ts)

    # Create observer location, using the correct keyword 'elevation_m'
    observer = wgs84.latlon(latitude, longitude, elevation_m=elevation)

    # Time range for predictions
    t0 = ts.now()
    t1 = ts.from_datetime(t0.utc_datetime() + timedelta(days=days_ahead))

    # Find events (rise, culmination, set) with a minimum altitude of 10 degrees
    times, events = satellite.find_events(observer, t0, t1, altitude_degrees=10.0)

    # Process and display passes
    passes = []
    # Event codes: 0=rise, 1=culmination, 2=set
    i = 0
    while i < len(events) - 2:
        # A pass is a sequence of rise, culmination, set
        if events[i] == 0 and events[i+1] == 1 and events[i+2] == 2:
            rise_time = times[i]
            culmination_time = times[i + 1]
            set_time = times[i + 2]

            # Calculate maximum elevation at culmination
            difference = satellite - observer
            topocentric = difference.at(culmination_time)
            alt, az, dist = topocentric.altaz()

            # **FIXED**: Calculate duration by subtracting Time objects directly.
            # The result is in days, so we convert to minutes.
            duration = (set_time - rise_time) * 24 * 60

            pass_info = {
                'rise_time': rise_time.utc_datetime(),
                'culmination_time': culmination_time.utc_datetime(),
                'set_time': set_time.utc_datetime(),
                'max_elevation': alt.degrees,
                'duration_minutes': duration
            }
            passes.append(pass_info)

            # Print pass information
            print(f"\n--- Pass #{len(passes)} ---")
            print(f"  Rise:        {pass_info['rise_time'].strftime('%Y-%m-%d %H:%M:%S UTC')}")
            print(f"  Culmination: {pass_info['culmination_time'].strftime('%Y-%m-%d %H:%M:%S UTC')}")
            print(f"  Set:         {pass_info['set_time'].strftime('%Y-%m-%d %H:%M:%S UTC')}")
            print(f"  Max Elev:    {pass_info['max_elevation']:.1f}°")
            print(f"  Duration:    {pass_info['duration_minutes']:.1f} minutes")

            # Move index past this full pass
            i += 3
        else:
            # Move to the next event if the sequence isn't a full pass
            i += 1

    if not passes:
        if city_name:
             print(f"\nNo visible passes found for {city_name} in the next {days_ahead} days.")
        else:
             print(f"\nNo visible passes found in the next {days_ahead} days.")

    return passes


# --- Interactive Example Usage ---
# Define a dictionary of major cities and their coordinates
CITIES = {
    '1': {'name': 'New York', 'lat': 40.7128, 'lon': -74.0060},
    '2': {'name': 'London', 'lat': 51.5074, 'lon': -0.1278},
    '3': {'name': 'Tokyo', 'lat': 35.6762, 'lon': 139.6503},
    '4': {'name': 'Sydney', 'lat': -33.8688, 'lon': 151.2093},
    '5': {'name': 'Mumbai', 'lat': 19.0760, 'lon': 72.8777},
    '6': {'name': 'Bangalore', 'lat': 12.9716, 'lon': 77.5946},
    '7': {'name': 'Moscow', 'lat': 55.7558, 'lon': 37.6173}
}

# Prompt the user to select a city
print("--- Select a city to calculate upcoming ISS passes ---")
for key, city_info in CITIES.items():
    print(f"  [{key}] {city_info['name']}")

# Use try-except block to handle potential non-interactive environments
try:
    choice = input("\nEnter the number of your choice (e.g., '1' for New York): ")
    # Validate choice and run calculation
    if choice in CITIES:
        selected_city = CITIES[choice]
        # Call the corrected function with the selected city's data
        passes = calculate_iss_passes(
            latitude=selected_city['lat'],
            longitude=selected_city['lon'],
            city_name=selected_city['name']
        )
    else:
        print("\nInvalid choice. Defaulting to New York.")
        # Default to a city if input is invalid
        selected_city = CITIES['1']
        passes = calculate_iss_passes(
            latitude=selected_city['lat'],
            longitude=selected_city['lon'],
            city_name=selected_city['name']
        )
except (EOFError, RuntimeError):
    # Handle cases where input() is not supported (e.g., non-interactive script execution)
    print("\nNo user input detected. Defaulting to New York.")
    selected_city = CITIES['1']
    passes = calculate_iss_passes(
        latitude=selected_city['lat'],
        longitude=selected_city['lon'],
        city_name=selected_city['name']
    )
```

```python
# [REVISED CODE FOR TASK 4 CELL]

def plot_iss_ground_track(duration_minutes=90):
    """Plot ISS ground track for specified duration, with robust image handling."""

    # Load timescale and create satellite object
    ts = load.timescale()
    satellite = EarthSatellite(tle_line1, tle_line2, iss_name, ts)

    # Generate time points for the next `duration_minutes`
    now = ts.now()
    # Create a sequence of Time objects for the next N minutes
    times_array = ts.utc(now.utc.year, now.utc.month, now.utc.day, now.utc.hour,
                        now.utc.minute + np.arange(duration_minutes))

    # Calculate ISS subpoints (latitude and longitude)
    geocentric = satellite.at(times_array)
    subpoints = wgs84.subpoint(geocentric)
    lats = subpoints.latitude.degrees
    lons = subpoints.longitude.degrees

    # Create plot
    fig, ax = plt.subplots(figsize=(15, 7))
    ax.set_title(f'ISS Ground Track - Next {duration_minutes} minutes')
    ax.set_xlabel('Longitude (°)')
    ax.set_ylabel('Latitude (°)')
    ax.set_xlim(-180, 180)
    ax.set_ylim(-90, 90)
    ax.grid(True, linestyle='--', alpha=0.5)

    # --- START OF FIX ---
    # Attempt to load a world map image as the background
    try:
        from PIL import Image, UnidentifiedImageError
        import io

        img_url = "https://upload.wikimedia.org/wikipedia/commons/8/83/Equirectangular_projection_SW.jpg"

        # Add a User-Agent header to mimic a browser request
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
        }

        response = requests.get(img_url, headers=headers)
        response.raise_for_status()  # Ensure the download was successful

        img = Image.open(io.BytesIO(response.content))
        ax.imshow(img, extent=[-180, 180, -90, 90], aspect='auto', alpha=0.6)

    except (ImportError, requests.RequestException, UnidentifiedImageError) as e:
        # Catch multiple potential errors: Pillow not installed, network error, or invalid image data
        print(f"Warning: Could not load world map background image. Reason: {e}")
        print("Displaying track on a blank grid.")
    # --- END OF FIX ---

    # Split the track into segments to handle plotting across the antimeridian (-180/180 line)
    path_segments = []
    current_segment_lons = [lons[0]]
    current_segment_lats = [lats[0]]

    for i in range(len(lons) - 1):
        # A large jump in longitude indicates a wrap-around
        if abs(lons[i+1] - lons[i]) > 180:
            path_segments.append((current_segment_lons, current_segment_lats))
            current_segment_lons = [lons[i+1]]
            current_segment_lats = [lats[i+1]]
        else:
            current_segment_lons.append(lons[i+1])
            current_segment_lats.append(lats[i+1])
    path_segments.append((current_segment_lons, current_segment_lats))

    # Plot each segment
    for seg_lons, seg_lats in path_segments:
        ax.plot(seg_lons, seg_lats, 'r-', linewidth=2)

    # Mark current and final positions
    ax.plot(lons[0], lats[0], 'yo', markersize=10, markeredgecolor='black', label='Current Position')
    ax.plot(lons[-1], lats[-1], 'go', markersize=8, markeredgecolor='black', label=f'Position in {duration_minutes} min')

    ax.legend()
    plt.tight_layout()
    plt.show()

    return lats, lons

# Plot the ground track
ground_track_data = plot_iss_ground_track(90)
```

```python
def predict_multiple_locations(locations_file='locations.csv', results_file='iss_passes_results.csv'):
    """Predict ISS passes for multiple locations from a list and save to CSV."""

    # Sample locations data
    locations = [
        {'name': 'New York', 'lat': 40.7128, 'lon': -74.0060},
        {'name': 'London', 'lat': 51.5074, 'lon': -0.1278},
        {'name': 'Tokyo', 'lat': 35.6762, 'lon': 139.6503},
        {'name': 'Sydney', 'lat': -33.8688, 'lon': 151.2093},
        {'name': 'Mumbai', 'lat': 19.0760, 'lon': 72.8777},
        {'name': 'Bangalore', 'lat': 12.9716, 'lon': 77.5946}
    ]

    # Save locations to a CSV file (for reference)
    with open(locations_file, 'w', newline='') as f:
        writer = csv.DictWriter(f, fieldnames=['name', 'lat', 'lon'])
        writer.writeheader()
        writer.writerows(locations)
    print(f"Locations list saved to '{locations_file}'")

    # Process each location and collect results
    all_results = []

    print("\n--- Processing Multiple Locations ---")
    for location in locations:
        print(f"\nProcessing {location['name']}...")

        try:
            # We calculate passes for the next 2 days to increase chances of finding one
            passes = calculate_iss_passes(
                location['lat'],
                location['lon'],
                days_ahead=2
            )

            # For this task, we will just record the *next* visible pass
            if passes:
                first_pass = passes[0]
                all_results.append({
                    'Location': location['name'],
                    'Rise Time (UTC)': first_pass['rise_time'].strftime('%Y-%m-%d %H:%M'),
                    'Culmination Time (UTC)': first_pass['culmination_time'].strftime('%Y-%m-%d %H:%M'),
                    'Set Time (UTC)': first_pass['set_time'].strftime('%Y-%m-%d %H:%M'),
                    'Max Elevation (deg)': f"{first_pass['max_elevation']:.1f}",
                    'Duration (min)': f"{first_pass['duration_minutes']:.1f}"
                })
            else:
                all_results.append({
                    'Location': location['name'],
                    'Rise Time (UTC)': 'No visible passes in next 2 days',
                    'Culmination Time (UTC)': '-', 'Set Time (UTC)': '-',
                    'Max Elevation (deg)': '-', 'Duration (min)': '-'
                })

        except Exception as e:
            print(f"Error processing {location['name']}: {e}")

    # Save results to the output CSV file
    if all_results:
        with open(results_file, 'w', newline='') as f:
            fieldnames = all_results[0].keys()
            writer = csv.DictWriter(f, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(all_results)
        print(f"\n--- Batch prediction complete. Results saved to '{results_file}' ---")

    return all_results

# Run batch predictions
batch_results = predict_multiple_locations()
# To display results neatly in a notebook, you could use pandas
# import pandas as pd
# display(pd.DataFrame(batch_results))
```

```python
def create_interactive_tracker():
    """Create an interactive ISS pass tracker using ipywidgets."""

    # Define preset locations
    locations = {
        'New York': (40.7128, -74.0060),
        'London': (51.5074, -0.1278),
        'Tokyo': (35.6762, 139.6503),
        'Sydney': (-33.8688, 151.2093),
        'Mumbai': (19.0760, 72.8777),
        'Bangalore': (12.9716, 77.5946),
        'Custom': (0.0, 0.0) # Placeholder for custom input
    }

    # Create widgets for user input
    location_dropdown = widgets.Dropdown(
        options=list(locations.keys()),
        value='Bangalore',
        description='Location:'
    )

    lat_input = widgets.FloatText(value=locations['Bangalore'][0], description='Latitude:', disabled=True)
    lon_input = widgets.FloatText(value=locations['Bangalore'][1], description='Longitude:', disabled=True)

    days_slider = widgets.IntSlider(value=3, min=1, max=10, step=1, description='Days:')

    calculate_button = widgets.Button(description='Calculate Passes', button_style='success', icon='satellite')

    # Output widget to display results
    output = widgets.Output()

    # Function to update lat/lon fields when dropdown changes
    def update_coordinates_on_change(change):
        if location_dropdown.value == 'Custom':
            lat_input.disabled = False
            lon_input.disabled = False
        else:
            lat_input.disabled = True
            lon_input.disabled = True
            lat, lon = locations[location_dropdown.value]
            lat_input.value = lat
            lon_input.value = lon

    # Function to run calculations when the button is clicked
    def calculate_passes_on_click(b):
        with output:
            clear_output(wait=True) # Clear previous output
            loc_name = location_dropdown.value
            print(f"Calculating passes for {loc_name} ({lat_input.value:.4f}, {lon_input.value:.4f})...")
            print(f"Searching for the next {days_slider.value} days.")
            try:
                calculate_iss_passes(
                    lat_input.value,
                    lon_input.value,
                    days_ahead=days_slider.value
                )
            except Exception as e:
                print(f"An error occurred: {e}")

    # Link widget events to functions
    location_dropdown.observe(update_coordinates_on_change, names='value')
    calculate_button.on_click(calculate_passes_on_click)

    # Arrange and display the UI
    ui = widgets.VBox([
        widgets.HTML("<h2>Interactive ISS Pass Predictor</h2>"),
        location_dropdown,
        widgets.HBox([lat_input, lon_input]),
        days_slider,
        calculate_button,
        output
    ])
    display(ui)

# Create and display the interactive interface
create_interactive_tracker()
```

```python
def analyze_orbital_drift(days_to_analyze=30):
    """Analyze and plot simulated changes in ISS orbital elements over time."""

    print(f"--- Analyzing Simulated Orbital Drift over {days_to_analyze} Days ---")
    print("NOTE: This uses simulated data. In a real scenario, you would download historical TLEs.")

    dates, inclinations, eccentricities, periods = [], [], [], []

    # Base values typical for the ISS
    base_inclination = 51.64
    base_eccentricity = 0.0007
    base_mean_motion = 15.49  # Revolutions per day

    # Simulate daily variations over the past number of days
    np.random.seed(42) # for reproducible results
    for day in range(days_to_analyze):
        # Go back in time for the plot
        date = datetime.utcnow() - timedelta(days=days_to_analyze - day - 1)
        dates.append(date)

        # Simulate small random variations for inclination and eccentricity
        inc = base_inclination + np.random.normal(0, 0.001)
        ecc = base_eccentricity + abs(np.random.normal(0, 0.00005)) # Eccentricity is non-negative

        # Simulate orbital decay (mean motion increases) and occasional reboosts (sudden decrease)
        decay = (day * 0.00005) # Gradual decay
        reboost = -0.015 if day == 20 else 0 # Simulate a reboost on day 20
        mm = base_mean_motion + decay + reboost + np.random.normal(0, 0.0001)

        inclinations.append(inc)
        eccentricities.append(ecc)
        periods.append((24 * 60) / mm) # Orbital period in minutes

    # Create visualization
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 12), sharex=True)
    fig.suptitle('Simulated ISS Orbital Element Drift', fontsize=16)

    ax1.plot(dates, inclinations, 'b.-', label='Inclination')
    ax1.set_ylabel('Inclination (°)')
    ax1.grid(True, alpha=0.5)

    ax2.plot(dates, eccentricities, 'g.-', label='Eccentricity')
    ax2.set_ylabel('Eccentricity')
    ax2.grid(True, alpha=0.5)

    ax3.plot(dates, periods, 'r.-', label='Period')
    ax3.set_ylabel('Orbital Period (min)')
    ax3.set_xlabel('Date')
    ax3.grid(True, alpha=0.5)

    # Improve formatting
    fig.autofmt_xdate(rotation=45)
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

    # Interpretation
    print("\n--- Analysis of Trends ---")
    print("1. Inclination: Remains very stable, as it requires significant energy to change.")
    print("2. Eccentricity: Stays very low (near-circular orbit) with minor fluctuations.")
    print("3. Orbital Period: Shows a general downward trend, corresponding to altitude loss from atmospheric drag. The sharp jump simulates a 'reboost' maneuver, where the ISS fires its thrusters to regain altitude and speed, increasing its orbital period.")

    return dates, inclinations, eccentricities, periods

# Run the orbital drift analysis
orbital_data = analyze_orbital_drift(30)
```
